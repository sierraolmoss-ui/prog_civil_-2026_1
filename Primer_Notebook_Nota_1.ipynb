## Problema de Ingeniería

### Cálculo del **área de un terreno rectangular**

Un ingeniero necesita calcular el **área de un terreno** para estimar la cantidad de material o cultivo que puede colocarse.

La fórmula es:

[
Area = Largo \times Ancho
]

---

# 1. Definición del algoritmo (Entradas – Proceso – Salidas)

### Entradas

Son los datos que el usuario debe ingresar.

* largo del terreno (m)
* ancho del terreno (m)

### Proceso

Operación matemática que se realizará.

* multiplicar largo × ancho

### Salida

Resultado que el programa mostrará.

* área del terreno en metros cuadrados

---

# 2. Pseudocódigo

```
Inicio

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

Fin
```

---

# 3. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica de ingeniería

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 4. Ejemplo de ejecución

```
Ingrese el largo del terreno en metros: 20
Ingrese el ancho del terreno en metros: 10

El área del terreno es: 200 m²
```
