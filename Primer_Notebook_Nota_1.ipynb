## Problema de Ingeniería

### Cálculo del **área de un terreno rectangular**

Un ingeniero necesita calcular el **área de un terreno** para estimar la cantidad de material o cultivo que puede colocarse.

La fórmula es:

[
Area = Largo \times Ancho
]

---

# 1. Definición del algoritmo (Entradas – Proceso – Salidas)

### Entradas

Son los datos que el usuario debe ingresar.

* largo del terreno (m)
* ancho del terreno (m)

### Proceso

Operación matemática que se realizará.

* multiplicar largo × ancho

### Salida

Resultado que el programa mostrará.

* área del terreno en metros cuadrados

---

# 2. Pseudocódigo

```
Inicio

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

Fin
```

---

# 3. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica de ingeniería

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

### Un ejemplo sencillo: El pensamiento algorítmico

Antes de lanzarnos a escribir código, recuerda que programar es **dar instrucciones claras**. Supongamos que queremos calcular el volumen de agua necesario para llenar un canal rectangular de riego.

**Lógica previa:**

1. Necesitamos la base del canal.
2. Necesitamos la altura del tirante de agua.
3. Necesitamos la longitud del tramo.
4. Multiplicamos los tres valores.

Aquí tienes un ejemplo de cómo se vería esto en Python, siguiendo **buenas prácticas** de nombres descriptivos:

````python
def calcular_volumen_canal(base_m, altura_m, longitud_m):
    """
    Calcula el volumen de un tramo de canal rectangular.
    """
    volumen_m3 = base_m * altura_m * longitud_m
    return volumen_m3

# Datos de entrada (Imagina que vienen de un levantamiento topográfico)
ancho_base = 1.5  # metros
tirante_agua = 0.8 # metros
largo_tramo = 100.0 # metros

# Llamada a la función
resultado = calcular_volumen_canal(ancho_base, tirante_agua, largo_tramo)

print(f"El volumen de agua en el tramo es: {resultado} m3")
````
