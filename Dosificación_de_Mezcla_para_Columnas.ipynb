# **Dosificación de Mezcla para Columnas**

**Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

 Se sabe que para un concreto estándar de 210 kg/cm2  (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v= π * r^2 * h$$

(Sugerencia: usar 3.1416
 o importar math.pi).

4. Calcular la cantidad de sacos:$$sacos = v * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):


Volumen ≈0.377 m3

Sacos necesarios ≈3.58


(el estudiante verá aquí la importancia de los tipos decimales).

```markdown
Aquí tienes la explicación de cada línea del código:

```python
h = float(input("Ingrese la altura de la columna en metros: "))
```
Esta línea solicita al usuario que introduzca la altura de la columna en metros. La función `input()` lee la entrada como una cadena de texto, y `float()` la convierte en un número decimal (flotante) para poder realizar cálculos con ella. El valor se almacena en la variable `h`.

```python
d = float(input("Ingrese el diámetro de la columna en metros: "))
```
Similar a la línea anterior, esta línea solicita al usuario el diámetro de la columna en metros. La entrada se convierte a un número flotante y se guarda en la variable `d`.

```python
r = d/2
```
Esta línea calcula el radio de la columna. El radio (`r`) es la mitad del diámetro (`d`). El resultado se almacena en la variable `r`.

```python
v = 3.1416 * r**2 * h
```
Esta línea calcula el volumen (`v`) de la columna circular utilizando la fórmula $$v = \pi * r^2 * h$$. Se usa `3.1416` como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y `h` es la altura. El resultado se guarda en la variable `v`.

```python
sacos = v * 9.5
```
Esta línea calcula la cantidad de sacos de cemento necesarios. Se multiplica el volumen calculado (`v`) por la constante `9.5`, que representa los sacos de cemento requeridos por cada metro cúbico de mezcla. El resultado se almacena en la variable `sacos`.

```python
print(f"Volumen ≈{v:.3f} m3")
```
Esta línea imprime el volumen calculado de la columna. Usa una f-string para formatear la salida, mostrando el volumen (`v`) con tres decimales (`.3f`) seguido de la unidad " m3".

```python
print(f"Sacos necesarios ≈{sacos:.2f}")
```
Finalmente, esta línea imprime la cantidad de sacos de cemento necesarios. También usa una f-string para formatear la salida, mostrando el número de sacos (`sacos`) con dos decimales (`.2f`).
```

````python
# Entrada:Solicitar la altura de la columna en metros.
#Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# Proceso : Calcula el radio r , calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# Salida : imprimir el resultado
print(f"Volumen ≈{v:.3f} m3")
print(f"Sacos necesarios ≈{sacos:.2f}")
````
